### 산술 연산자

#### 숫자 연산

```python
num1 = 10
num2 = 7
# 나누기 /, 나머지 %, 몫 //
print('나누기 : ',num1/num2)
print('나머지 : ',num1%num2)
print('몫 : ',num1//num2)
```

```python
# 문자열을 체크하는 함수 : isalpha(), isdigit()
# 문자열을 전처리(깔끔하게 정리하는 작업) 할 때 사용 --> 조건문 활용

# isalpha() : 문자열이 문자인지 아닌지를 확인하는 함수
# 숫자가 '단 하나'라도 있다면 --> False 반환
# 한글도 문자로 인식 -> True 반환

# isidgit() : 문자열이 숫자인지 아닌지를 확인하는 함수
# 문자가 '단 하나'라도 있다면 --> False 반환
# 음수부호(-), 소수점(.)을 문자로 판단 --> 실수나 음수를 판단하지 못함. False 반환
# 오직 0을 포함한 양수형 정수, 지수 --> True 반환

str1.isdigit()
str1.isalpha()
```

### 비교 연산자

```python
# True, False로 결과가 출력되는 연산자
# >, >=, <, <=, ==(같다), !=(같지 않다)

a = 3
b = 7

print(a > b) # False
print(a <= b) # True
print(a == b) # False
print(a != b) # True
```

### 논리 연산자

```python
# not, and, or
# not : 논리를 뒤집어버리는 연산자 True -> False, False -> True
a, b = 5, 10
not a < b # True -> False
```

```python
# and, or
# and : 두 개의 논리를 연결하여 확인, 두 논리가 모두 참일 경우에만 True
# or : 한개의 논리만 참일 경우에도 True

print(3 > 5 and 10 == 10) # False
print(3 > 5 or 10 == 10) # True
```

### 삼항 연산자
- True일때 실행문장 if 조건식 else False일때 실행문장

```python
# 60점 이상이면 합격, 미만이면 불합격 츨력해보기
score = 50

'합격' if score > 60 else '불합격'
```

```python
a = int(input('정수입력 >> '))
b = int(input('정수입력 >> '))
# print('a : {}'.format(a) if a > b else 'b : {}'.format(b))
print(f'a : {a}' if a > b else f'b : {b}')
```
